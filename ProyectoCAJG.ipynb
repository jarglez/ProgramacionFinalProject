# **1. Introducci√≥n**

## - ‚ö†Ô∏è Descripci√≥n del problema: ¬øQu√© queremos analizar o predecir?

El infarto agudo de miocardio (IAM) es una de las principales causas de mortalidad en el mundo. Durante la hospitalizaci√≥n, algunos pacientes desarrollan **complicaciones graves y mortales**. La capacidad de identificar los pacientes con riesgo de estas complicaciones desde el ingreso del paciente permitir√≠a **optimizar los recursos cl√≠nicos**, dirigir el **esfuerzo terape√∫tico**, mejorar la toma de decisiones y **reducir la mortalidad hospitalaria**.

En este an√°lisis, buscamos desarrollar un modelo de predicci√≥n que permita identificar qu√© pacientes tienen mayor riesgo de presentar **complicaciones mortales** tras un infarto.  

## - üéØ Objetivo del an√°lisis: ¬øQu√© impacto tendr√° este an√°lisis?

El objetivo principal de este estudio es construir un modelo de predicci√≥n de mortalidad y complicaciones mortales basado en datos cl√≠nicos y demogr√°ficos disponibles en el momento del ingreso, o en las primeras horas de estancia intrahospitalaria del paciente. Este modelo clasificar√° a los pacientes en una de las siguientes categor√≠as mortales seg√∫n su evoluci√≥n:

- **Vivo**: El paciente sobrevive sin complicaciones cr√≠ticas. (**NO TUVO COMPLICACIONES FATALES, PUDO HABER TENIDO COMPLICACIONES NO FATALES**)
- **Shock cardiog√©nico** ü´Ä: Estado de insuficiencia circulatoria grave debido a la incapacidad del coraz√≥n para bombear sangre de manera efectiva.
- **Edema pulmonar** üåä: Acumulaci√≥n de l√≠quido intersticial en los pulmones, lo que dificulta la oxigenaci√≥n y puede llevar a insuficiencia respiratoria.
- **Ruptura mioc√°rdica** üíî: Desgarro de la pared del miocardio, lo que puede causar un taponamiento card√≠aco y una r√°pida muerte.
- **Progresi√≥n de insuficiencia card√≠aca congestiva** üîÑ: Deterioro progresivo de la funci√≥n card√≠aca, lo que lleva a una acumulaci√≥n de l√≠quidos y falla multiorg√°nica y finalmente muerte.
- **Tromboembolismo** ü©∏: Formaci√≥n de co√°gulos sangu√≠neos que pueden desplazarse a √≥rganos vitales, causando infartos cerebrales, pulmonares o sist√©micos.
- **Asistolia** ‚õî: Ausencia total de actividad el√©ctrica en el coraz√≥n, lo que resulta en un paro card√≠aco irreversible.
- **Fibrilaci√≥n ventricular** ‚ö°: Ritmo card√≠aco ca√≥tico que impide que el coraz√≥n bombee sangre de manera efectiva, llevando r√°pidamente a la muerte si no se trata de inmediato.

Este modelo permitir√≠a al personal de salud **identificar de manera temprana** a los pacientes con mayor riesgo de complicaciones y tomar medidas preventivas, como **ajustes en el tratamiento, monitoreo intensivo y estrategias de intervenci√≥n temprana**. En √∫ltima instancia, esta herramienta podr√≠a **reducir la mortalidad hospitalaria y mejorar los resultados cl√≠nicos** en pacientes con infarto agudo de miocardio.

## - ‚ö†Ô∏è Debilidades del Dataset  

A pesar de que la **base de datos de complicaciones del infarto de miocardio** es una fuente valiosa de informaci√≥n cl√≠nica, presenta varias limitaciones que deben considerarse al utilizarla en an√°lisis modernos:  

### üìÖ **Fecha de recolecci√≥n de los datos (1992-1995)**  
- Los datos fueron recopilados hace **m√°s de 25 a√±os**, lo que plantea un desaf√≠o en cuanto a su **relevancia actual**.  
- Desde entonces, ha habido **avances significativos** en el diagn√≥stico, tratamiento y manejo del infarto agudo de miocardio, lo que podr√≠a hacer que los patrones de complicaciones en la base de datos **no reflejen la realidad m√©dica actual**.  
- Medicamentos, procedimientos como **avancees en reanimaci√≥n cardiopulmonar**, **detecci√≥n temprana de isquemia cardiaca**, **dispositivos como desfirbiladores autom√°ticos externos**, **programas como c√≥digo infarto**, la **angioplastia primaria** y el uso de **stents**, as√≠ como el manejo en unidades de cuidados intensivos, han evolucionado y cambiado la **tasa de complicaciones**. Por ejemplo, los datos de esta fuente no clasifican el infarto con el paradigma moderno con o sin elevaci√≥n del segmento ST, lo que disminuye su utilidad. Otro ejemplo es el tratamiento fibrinol√≠tico que usan son del grupo de la esptreptoquinada, tratamiento que actualmente no est√° en uso.
- A pesar de contar con documentos que describen bien las variables de carece de la descripci√≥n metodol√≥gica de como se recolectaron los datos y la raz√≥n por lo que existen valores nulos o perdidos.  

### üåç **Ubicaci√≥n geogr√°fica: Rusia en un periodo de crisis**  
- La base de datos fue recolectada en **Krasnoyarsk, Rusia**, lo que podr√≠a **limitar su aplicabilidad** a otros contextos sanitarios y poblaciones con diferentes caracter√≠sticas gen√©ticas, ambientales y socioecon√≥micas.  
- **Impacto del colapso de la Uni√≥n Sovi√©tica** (1991):  
  - Rusia atravesaba una **grave crisis econ√≥mica y sanitaria** en los a√±os 90.  
  - Escasez de insumos m√©dicos, tecnolog√≠a limitada y condiciones hospitalarias precarias podr√≠an haber afectado la **calidad del tratamiento y la recopilaci√≥n de datos**.  
  - Falta de acceso a medicamentos modernos y procedimientos avanzados podr√≠a haber influido en la **tasa de complicaciones** registradas.  

### üìâ **Valores faltantes en los datos (7.6%)**  
- La base de datos presenta un **7.6% de valores faltantes**, lo que puede afectar la **precisi√≥n y confiabilidad** del an√°lisis.  
- Se requerir√° aplicar **m√©todos de imputaci√≥n o eliminaci√≥n** para manejar estos datos de manera adecuada.  
- Tambi√©n se observa la ausencia de algunas columnas originales en el dataset, no solo por su numeraci√≥n, sino tambi√©n por los hallazgos. Por ejemplo, en los bloqueos falta el de **2¬∫ grado**. Asimismo, se identifican las siguientes ausencias en los grupos de variables:

````python
# Buscamos ouliers en edad por el metodo gr√°rfico
sns.boxplot(y=df["edad"])

plt.title("Boxplot de la edad para buscar outliers")
plt.show()
````

Notamos que con el m√©todo gr√°fico existen outliers.

````python
# Con cuartiles

Q1 = df["edad"].quantile(0.25)
Q3 = df["edad"].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df[(df["edad"] < limite_inferior) | (df["edad"] > limite_superior)]

outliers
````

Confirmamos que existen de nuevo con un seguindo m√©todo, por loq ue vamos a descartar estos tres registros.

````python
df = df.drop(index=outliers.index)

df.info()
````
